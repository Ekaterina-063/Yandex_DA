```python
real_estate['price_of_square_meter'] = real_estate['last_price']/real_estate['total_area'] 
```

–°—Ç–æ–ª–±–µ—Ü —Å –¥–Ω—ë–º –Ω–µ–¥–µ–ª–∏ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è:

```python
real_estate['weekday'] = real_estate['first_day_exposition'].dt.weekday 
real_estate['weekday'].unique() # –ø—É–±–ª–∏–∫–∞—Ü–∏—è –æ—Å—É—â–µ—Å—Ç–≤–ª—è–µ—Ç—Å—è –µ–∂–µ–¥–Ω–µ–≤–Ω–æ —Å –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫–∞(0) –ø–æ –≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ (6)
```

–°—Ç–æ–ª–±–µ—Ü —Å –º–µ—Å—è—Ü–µ–º –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è:

```python
real_estate['month'] = real_estate['first_day_exposition'].dt.month
real_estate['month'].unique() # –ü—É–±–ª–∏–∫–∞—Ü–∏—è –æ—Å—É—â–µ—Å—Ç–≤–ª—è–µ—Ç—Å—è –µ–∂–µ–º–µ—Å—è—á–Ω–æ —Å —è–Ω–≤–∞—Ä—è(1) –ø–æ –¥–µ–∫–∞–±—Ä—å(12)
```

–°—Ç–æ–ª–±–µ—Ü —Å –≥–æ–¥–æ–º –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è:

```python
real_estate['year'] = real_estate['first_day_exposition'].dt.year
real_estate['year'].unique() # –ø–µ—Ä–µ–¥ –Ω–∞–º–∏ –¥–∞–Ω–Ω—ã–µ –∑–∞ –ø–µ—Ä–∏–æ–¥ —Å 2015 –ø–æ 2019 –≥–æ–¥—ã –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ
```

–°—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏–µ–π —ç—Ç–∞–∂–Ω–æ—Å—Ç–∏:

```python
def categorize_floor(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return '–ø–µ—Ä–≤—ã–π'
    if floor == floors_total:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    if floor > 1 and floor != floors_total:
        return'–¥—Ä—É–≥–æ–π'
real_estate['floor_category'] = real_estate.apply(categorize_floor, axis=1) 
```

C—Ç–æ–ª–±–µ—Ü —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≥–æ—Ä–æ–¥–∞ –≤ –∫–º:

```python
real_estate['cityCenters_nearest_km'] = (real_estate['cityCenters_nearest']/1000).round(0).astype(int) 
```
